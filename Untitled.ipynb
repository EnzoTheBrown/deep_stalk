```python
from polygon import RESTClient
API_KEY = 'fGEPZ3b3JZjew8nxOW2YeWXl2N7GZ7cZ'

client = RESTClient(API_KEY)

start_date = '2020-12-14'
end_date = '2022-12-15'
URL = ('https://api.polygon.io/v2/aggs/ticker/{company}/range/1/day/{start_date}'
       '/{end_date}?adjusted=true&sort=asc&limit=120&apiKey={API_KEY}')


companies = [
       "LVMHF",
       "AIQUF",
       "LRLCF",
       "TTFNF",
]

payloads = []

for company in companies:
       print(company)
       aggs = client.get_aggs(
           company,
           1,
           "day",
           start_date,
           end_date,
       )
       payloads += [{'company': company, 'timestamp': a.timestamp, 'value': a.close} for a in aggs]

import pandas as pd

df = pd.DataFrame(payloads)

import matplotlib.pyplot as plt

from datetime import datetime

df['datetime']  = df.apply(lambda row: datetime.fromtimestamp(row['timestamp']/1000), axis=1)
```

```python
df['mean_30'] = df.groupby('company')['value'].transform(lambda x: x.rolling(100, 1).mean())
df['std_30'] = df.groupby('company')['value'].transform(lambda x: x.rolling(100, 1).std())
df['pos_30'] = df['mean_30'] + df['std_30']
df['neg_30'] = df['mean_30'] - df['std_30']
df['2_pos_30'] = df['mean_30'] + 2*df['std_30']
df['2_neg_30'] = df['mean_30'] - 2*df['std_30']
```

```python
for company in companies:
    print(company)
    print(df['value'].iloc[-1] < df['2_neg_30'].iloc[-1])
    dd = df[df['company'] == company]
    plt.plot(dd['value'])
    plt.plot(dd['mean_30'])
    plt.plot(dd['pos_30'])
    plt.plot(dd['neg_30'])
    plt.plot(dd['2_pos_30'])
    plt.plot(dd['2_neg_30'])
    plt.show()
```
